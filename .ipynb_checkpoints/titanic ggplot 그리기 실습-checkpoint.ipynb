```python
import pandas as pd
import numpy as np
from plotnine import *
```

```python
train = pd.read_csv('train.csv')
```

```python
train.head()
```

```python
train = train.dropna()
```

```python
train['Survived'] = train['Survived'].astype(int)
train['Pclass'] = train['Pclass'].astype('category')
train['Age'] = train['Age'].astype(int)
```

```python
(ggplot(train)
 + aes(x='Age',y='Survived', fill='Sex')
 + geom_col()
)
```

```python
train.dtypes
```

```python
(ggplot(train)
 + aes(x='Sex', y='Age')
 + geom_boxplot(color=['green', 'red'])
)
```

```python
train['Survived'].dtypes
```

```python
train.groupby(['Pclass', 'Survived']).mean()
```
